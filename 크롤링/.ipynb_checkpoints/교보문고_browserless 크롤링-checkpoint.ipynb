# browserless 크롤링

- 앞선 selenium은 물리적으로 직접 브라우저를 켜 놓고 크롤링을 했습니다.
- 이렇게 크롤링을 하면 사람과 컴퓨터를 구분할 수 없기 때문에 크롤링을 막는 사이트를 뚫기 유요하고
- 무엇보다 눈에 동작이 보여서 직관적이지만 실행속도가 느리다는 단점이 있습니다.

- selenium을 통한 크롤링 시에는 특히 비동기 요청이나 api 크롤링을 하는데 어려움이 있기 때문에
- 이를 보완하기 위해 requests를 사용합니다.

```python
import requests # selenium과 달리 내장되어있어서 따로 설치할 필요없음
import codecs
from bs4 import BeautifulSoup
import time
```

```python
# 1. requests.get("접속주소") 로 접속
res = requests.get("http://www.naver.com")
# 2. 받아온 변수 res.text를 하면 해당 페이지의 소스코드가 나옵니다.
source = res.text
source
```

```python
# 위 내용을 토대로 교보문고 순회 크롤러를 이용해 1~200위까지의 도서 제목, 가격, 저자를 얻어와서
# txt나 csv파일로 만들어주세요
```

```python
# 전체 제목, 작가, 가격 정보를 저장하기 위한 리스트 생성
total_title = []
total_author = []
total_price = []

for i in range(1, 11):
    # 해당 페이지 접속 후 소스코드 파이썬 변수로 저장하기
    kyobo = requests.get("https://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?targetPage=%s" % i)
    source = kyobo.text
    # 저장된 파이썬 코드 파싱하기
    
    parsed_source = BeautifulSoup(source, "html.parser")
    parsed_source = parsed_source.find_all("ul", class_="list_type01")[0]
    
    div_title_list = parsed_source.find_all("div", class_="title")
    div_author_list = parsed_source.find_all("div", class_="author")
    div_price_list = parsed_source.find_all("div", class_="price")
    
    for item in div_title_list:
        total_title.append(item.text.strip())
    for item in div_author_list:
        total_author.append(item.text.strip().replace("\t", "").replace("\r\n", "").replace("\n", ""))
    for item in div_price_list:
        total_price.append(item.text.strip().replace("\t", "").replace("\r\n", "").replace("\n", "").replace(",", ""))
    
    time.sleep(2)
```

```python
# 위에 얻어놓은 total_title, total_author, total_price를 이용해 csv형태로 데이터를 저장해보세요.

f = codecs.open("c:/crawler1/kyobo.txt", mode='w', encoding="utf-8")
f.write("책제목,책저자,책가격\n")

for i in range(len(total_title)):
    f.write(total_title[i] + ",")
    f.write(total_author[i] + ",")
    f.write(total_price[i] + "\n")
    # f.write(%s,%s,%s\n" % (total_title[i], total_author[i], total_price[i]))
f.close()
```
