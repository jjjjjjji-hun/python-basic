# 영화 평점 분석 실습

```python
# 라이브러리 불러오기
import pandas as pd
```

## 1. 영화 평점 데이터 적재 및 결합

```python
# 사용자 정보 적재
# 파일 이름 : 'data/movie_users.csv'
users = pd.read_csv('data/movie_users.csv')
users.drop('Unnamed: 0', axis=1, inplace=True)
users.head()
```

```python
# movie_details.csv 파일을 movie 변수에 입력해주세요.
# 그리고 쓸데없는 컬럼을 삭제해주시고, .head()로 조회해주세요.
movies = pd.read_csv('data/movie_details.csv')
movies.drop('Unnamed: 0', axis=1, inplace=True)
movies.head()
```

```python
# 영화 평점 정보 적재
# movie_rating.csv
# rating 변수에 저장해주세요.
ratings = pd.read_csv('data/movie_ratings.csv')
ratings.drop('Unnamed: 0', axis=1, inplace=True)
ratings.head()
```

```python
# 사용자수, 영화수, 평점수 개수를 print로 찍어주세요.
print(len(users))
print(len(movies))
print(len(ratings))
```

### 1. 데이터 결합
#### 1-1) 사용자 데이터와 영화 평점 데이터를 사용자아이디 컬럼 기준으로 공통으로 존재하는 데이터만 결합

```python
# 유저 정보의 컬럼 이름 확인
users.columns
```

```python
# 레이팅(평점입력) 정보의 컬럼 이름 확인
ratings.columns
```

```python
# users, rating의 배열 구조 확인
users.shape
```

```python
ratings.shape
```

```python
# data1 변수에 users, ratings 순으로 합쳐주세요.
data1 = users.merge(ratings)

# 실행 후 data1.shape로 배열 구조를 확인해주세요.
data1.shape
```

#### 1-2) 1번 결과에 영화 아이디 컬럼 기준으로 영화 정보 추가하기

```python
data1.head(3)
```

```python
# data1과 movies를 머지해서 data2변수에 저장해 영화정보를 마저 추가해주세요.
data2 = data1.merge(movies)

# .head()로 머지된 데이터를 간단히 확인해주세요.
data2.head()
```

### 2. 인기 있는 영화 찾기

- 기준 : 평점이 4.5 이상이고 평점 개수가 500개 이상인 영화

```python
# 평점이 높은 순서로 10개 정렬(head 활용)
data2.sort_values(by='평점', ascending=False).head(10)
```

```python
# 하나의 영화에 여러개의 평점이 존재하는 1:N 관계이므로
# 사용자 아이디는 사실상 영화별 평균평점을 구하는데는 불필요한 자료임
# 따라서 내가 원하는 로우(index) : 영화 제목
# 집계함수(aggfunc) : 평균
# 으로 pivot_table을 data2에 걸어 집계해보세요. 정렬 x
data3 = data2.pivot_table(index=('제목'), values='평점', aggfunc='mean')
```

```python
# data3를 평점 높은순으로 정렬해주세요.
data3.sort_values(by='평점', ascending=False)
```

```python
# 평점을 기준으로 정렬하되 평균평점만 출력하는게 아니라, 평점 갯수도 같이 출력
# data2에 대해서 pivot_table을 걸 때
# 로우 : 제목
# 값 : 평점
# 집계함수 : ['mean', 'count']
result = data2.pivot_table(index='제목', values='평점', aggfunc=['mean', 'count'])
```
